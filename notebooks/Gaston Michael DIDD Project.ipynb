```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
import pip
pip.main(["install", "openpyxl"])
```

```python
TEIS_Referral_Data = pd.read_excel('../data/TEIS Referral Data for NSS pulled 7-1-22.xlsx')
TEIS_Referral_Data
```

```python
Inactive_TEIS_Data = TEIS_Referral_Data[TEIS_Referral_Data['Active']== 'I']
Inactive_TEIS_Data
```

```python
Inactive_TEIS_Data.info()
```

```python
Inactive_TEIS_Data
```

```python
Exit_Reasons_POE = Inactive_TEIS_Data.groupby(['POE','County SES ','Exit Reason'])['Child ID'].count()
Exit_Reasons_POE
```

```python
Exit_Reasons_POE = Exit_Reasons_POE.to_frame()
Exit_Reasons_POE.head()
```

```python
Exit_Reasons_POE = Exit_Reasons_POE.rename(columns = {'Child ID': 'Count'})
Exit_Reasons_POE.reset_index()
```

```python
Child_Phase_at_Exit = Inactive_TEIS_Data.groupby(['POE','Child Phase', 'Exit Reason'])['Child ID'].count()
Child_Phase_at_Exit
```

```python
Child_Phase_at_Exit = Child_Phase_at_Exit.to_frame()
Child_Phase_at_Exit
```

```python
Child_Phase_at_Exit = Child_Phase_at_Exit.reset_index()
```

```python
Child_Phase_at_Exit = Child_Phase_at_Exit.rename(columns = {'Child ID':'Count'})
```

```python
Child_Phase_at_Exit
```

```python
ET_Exit = Child_Phase_at_Exit[Child_Phase_at_Exit.POE=='ET']
ET_Exit
```

```python
#Shows the Final Child Phase and the exit reasons for ET
ET_Exit['Percent of Total'] = ET_Exit.Count / ET_Exit.Count.sum()*100
ET_Exit = ET_Exit.sort_values(by='Percent of Total', ascending = False).reset_index()
ET_Exit
```

```python
#Shows the total count and % of Final Phase for ET
ET_Phase = ET_Exit.groupby(['Child Phase'])['Count'].sum()
ET_Phase = ET_Phase.to_frame()
ET_Phase['Percent of Total'] = ET_Phase.Count / ET_Phase.Count.sum()*100
ET_Phase = ET_Phase.sort_values('Percent of Total', ascending=False).reset_index()
ET_Phase
```

```python
# ET Final Phases
fig, ax = plt.subplots(figsize=(12,12))

sns.barplot(ET_Phase['Child Phase'], ET_Phase['Count'],
        label='ET Inactive Referalls: Final Phase',
        color = 'blue', order=ET_Phase.sort_values('Count',ascending=False)['Child Phase'])
plt.legend()
plt.style.use('ggplot')
plt.ylabel('Count of Referalls');
```

```python
#ET Exit Reasons in order to determine how many children exited the program successfully vs unsuccesfully 
ET_Exit_Reasons = ET_Exit.groupby(['Exit Reason'])['Count'].sum()
ET_Exit_Reasons = ET_Exit_Reasons.to_frame()
ET_Exit_Reasons['Percent of Total'] = ET_Exit_Reasons.Count / ET_Exit_Reasons.Count.sum()*100
ET_Exit_Reasons = ET_Exit_Reasons.sort_values('Percent of Total', ascending=False).reset_index()
ET_Exit_Reasons
```

```python
# ET Exit Reasons
fig, ax = plt.subplots(figsize=(12,12))

sns.barplot(y=ET_Exit_Reasons['Exit Reason'], x=ET_Exit_Reasons['Count'],
        label='ET Inactive Referalls: Exit Reason',
        color = 'red', order=ET_Exit_Reasons.sort_values('Count',ascending=False)['Exit Reason'], orient='h')
plt.legend()
plt.style.use('ggplot')
plt.ylabel('Exit Reason');
```
