```python
Final_Phase = Inactive_TEIS_Data.groupby(['Child Phase'])['Child ID'].count()
Final_Phase = Final_Phase.to_frame().reset_index()
Final_Phase = Final_Phase.rename(columns = {'Child ID':'Count'})
Final_Phase['Percent of Total'] = Final_Phase.Count / Final_Phase.Count.sum()*100
Final_Phase
```

```python
plt.bar(Final_Phase['Child Phase'], Final_Phase['Count'],
        label='Inactive Referalls',
        color = 'blue')
plt.legend()
plt.style.use('ggplot')
plt.ylabel('Count of Referalls');
```
